# Trends & models

### Calculate P-model 

```python
for source_id in models:
    cmip_label = 'historical_%s_gs'% source_id
    print(cmip_label)
    co2_v = 'CO2_historical'
    stressB = 1
    model_tag ='sunlit_ww'
    variant = [model_tag, cmip_label, co2_v, stressB]

    df_ts_h = add_p_model_CMIP(df_ts_h, variant, common_mean, source_id=source_id)
```
